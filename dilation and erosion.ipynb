## Thresholding with trackbar

```python
import cv2
# img=cv2.imread("./Images/smarties.jpeg",0)
img=cv2.imread("./Images/avatar.png")
def nothing(x):
    print(x)
cv2.namedWindow("thresolding")
cv2.createTrackbar("trackbar","thresolding",128,255,nothing)
while(1):
    pos=cv2.getTrackbarPos("trackbar","thresolding")
    ret,th1=cv2.threshold(img,pos,255,cv2.THRESH_BINARY)
    ret,th2=cv2.threshold(img,pos,255,cv2.THRESH_BINARY_INV)
    ret,th3=cv2.threshold(img,pos,255,cv2.THRESH_TRUNC)
    ret,th4=cv2.threshold(img,pos,255,cv2.THRESH_TOZERO)
    ret,th5=cv2.threshold(img,pos,255,cv2.THRESH_TOZERO_INV)
    cv2.imshow("thresolding",img)
    cv2.imshow("BINARY",th1)
    cv2.imshow("BINARY inverse",th2)
    cv2.imshow("Truncated",th3)
    cv2.imshow("TOZERO thresolding",th4)
    cv2.imshow("TOZERO thresolding inverse",th5)
    k=cv2.waitKey(1)
    if(k==27):
        break
cv2.destroyAllWindows()
```

## subploting of thresolding

```python
import matplotlib.pyplot as plt
im=cv2.imread("./Images/gradient.jpeg",0)
th=100
ret,th1=cv2.threshold(im,th,255,cv2.THRESH_BINARY)
ret,th2=cv2.threshold(im,th,255,cv2.THRESH_BINARY_INV)
ret,th3=cv2.threshold(im,th,255,cv2.THRESH_TRUNC)
ret,th4=cv2.threshold(im,th,255,cv2.THRESH_TOZERO)
ret,th5=cv2.threshold(im,th,255,cv2.THRESH_TOZERO_INV)
labels=["original img","binary","binary inverse","Truncated","ToZERO","TOZERO inverse"]
images=[im,th1,th2,th3,th4,th5]
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(labels[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

## Erosion , dilation,opening,closing

```python
import cv2
import numpy as np 
img=cv2.imread("./Images/smarties.jpeg",0)
ret,mask=cv2.threshold(img,193,255,cv2.THRESH_BINARY_INV)
kernel=np.ones((3,3),np.uint8)
ero=cv2.erode(mask,kernel,iterations=3)
dilate=cv2.dilate(mask,kernel,iterations=3)
cv2.imshow("original",img)
cv2.imshow("mask",mask)
cv2.imshow("erosion image",ero)
cv2.imshow("dilation immage",dilate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
opening=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)
closing=cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel)
cv2.imshow("original",img)
cv2.imshow("mask",mask)
cv2.imshow("erosion image",ero)
cv2.imshow("dilation immage",dilate)
cv2.imshow("opening",opening)
cv2.imshow("closing",closing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
j_img=cv2.imread("./Images/j.png",0)
ret,mask=cv2.threshold(j_img,127,255,cv2.THRESH_BINARY_INV)
kernel=np.ones((3,3),np.uint8)
j_erosion=cv2.erode(mask,kernel,iterations=2)
j_dilation=cv2.dilate(mask,kernel,iterations=2)
j_open=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)
j_close=cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel)
mg=cv2.morphologyEx(j_img,cv2.MORPH_GRADIENT,kernel)
cv2.imshow("original",j_img)
cv2.imshow("mask",mask)
cv2.imshow("erosion image",j_erosion)
cv2.imshow("dilation immage",j_dilation)
cv2.imshow("gradient",mg)
cv2.imshow("opening",j_open)
cv2.imshow("closing",j_close)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
