# Collage Maker

```python
import tkinter as tk
from tkinter.filedialog import askopenfile
import tkinter.filedialog as fd
import cv2
import numpy as np
from skimage import io
from PIL import Image, ImageTk
a=""
#only getting the file name 
a2=[]
a3=[]
#after reading all photos in np array format 
read_img=[]
resized=[]
def print_text():
    global text
    num=""
    num+=text.get()
    output.config(text=num)    
    text.delete(0,tk.END)
#     browse_files()
def browse_files():
    global a ,a1,a2
    file = fd.askopenfilenames(parent=wind, title='Choose a File')
    a=wind.splitlist(file) 
    #list of files uploaded 
    a1=list(a)
    for i in a1:
        a2.append(i[::-1])
    for j in a2:
        x=j.index("/")
        z=j[0:x]
        z=z[::-1]
        a3.append(z)    
    for i,j in enumerate(a3) :        
        output=tk.Label(wind,text=j)
        output.pack()
    output=tk.Label(wind,text="are selected")
    output.pack()
def get_output():
    global read_img,resized
    
    for i in a1:
        img=cv2.imread(i)
        read_img.append(img)
    for i in read_img:
        res=cv2.resize(i,(256,256))
        resized.append(res)
    if (len(read_img)%2==0):
        np1=np.hstack(resized[i] for i in range(0,len(read_img)//2))
        np2=np.hstack(resized[i] for i in range(len(read_img)//2,len(read_img)))
        ver=np.vstack([np1,np2])
        blue,green,red = cv2.split(ver)
        img = cv2.merge((red,green,blue))
        im = Image.fromarray(img)
        imgtk = ImageTk.PhotoImage(image=im)
        output2.configure(image=imgtk)
        output2.image=imgtk
    else:
        np1=np.hstack(resized[i] for i in range(0,len(read_img)//2))
        np2=np.hstack(resized[i] for i in range(len(read_img)//2,len(read_img)-1))
        np3=np.hstack(resized[i] for i in range(len(read_img)-1,len(read_img)))
        ver=np.vstack([np1,np2,np3])
        blue,green,red = cv2.split(ver)
        img = cv2.merge((red,green,blue))
        im = Image.fromarray(img)
        imgtk = ImageTk.PhotoImage(image=im)
        output2.configure(image=imgtk)
        output2.image=imgtk
        a=""
        #only getting the file name 
        a2=[]
        a3=[]
        #after reading all photos in np array format 
        read_img=[]
        resized=[]
        
        

    
wind=tk.Tk()
wind.geometry("720x720")
wind.title("Collage maker")
get=tk.Label(wind,text="Give number of photos for collage")
text=tk.Entry()
button=tk.Button(wind,text="output",width=25,command=get_output)
button1=tk.Button(wind,text="choose files",width=25,command=browse_files)
output=tk.Label(wind,text="")  
output2=tk.Label (wind,text="")

    
get.pack()
text.pack()
button1.pack()
button.pack()
output.pack()
output2.pack()



wind.mainloop()

```
